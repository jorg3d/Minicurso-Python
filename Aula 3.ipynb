   Aprendizado de máquina
   
   O aprendizado de máquina é uma área da Inteligência Artificial que busca o desenvolvimento de técnicas computacionais sobre o aprendizado e, também, a construção de sistemas que consigam adquirir conhecimento de maneira automática, sendo um sistema de aprendizado um programa de computador que toma decisões através de experiências acumuladas à partir de soluções bem sucedidas de problemas anteriores.(Monard & Baranauskas, 2003).
   Ou seja, aprendizagem de máquina estuda métodos computacionais para melhorar o desempenho, através da mecanização da aquisição de conhecimento baseado em experiências. (Langley & Simon, 1995)
    Dessa forma, podemos utilizar o aprendizado de máquina para solucionar diversos problemas, como por exemplo:
        -Processamento de Linguagem natural
        -Visão Computacional (processamento de imagens)
        -Análise de sequenciamento genético
        -Engenharia de Software
        -etc.


   Aprendizado Supervsionado e Não Supervisionado
    
   O aprendizado indutivo, que se efetua através de raciocínio sobre exemplos fornecidos por um processo externo ao sistema de aprendizado, pode ser dividido em dois subconjuntos, o supervisionado e o não-supervisionado. No aprendizado supervisionado, depende-se de um conjunto de exemplos de treinamento para serem fornecidos ao algoritmo de aprendizado, para que o mesmo possa realizar uma inferência posterior sobre novos exemplos.Já o não supervisionado, independe de um conjunto para treinamento, analisando os exemplos fornecidos, ele busca agrupá-los de alguma maneira, geralmente necessitando que, posteriormente, esses agrupamentos sejam analisados para definir o significado de cada agrupamento no contexto analisado. (Monard & Baranauskas, 2003)
   A partir disso, iremos abordar no curso dois algoritmos para cada tipo de aprendizado, realizando Clustering e Topic Modeling (Não-supervisionados). Já no supervisionado, utilizaremos Naive Bayes e Redes Neurais.


   Scikit Learn
    
   Scikit-Learn é um módulo em Python que integra uma vasta gama de algoritmos de aprendizado de máquina para problemas supervisionados ou não-supervisionados de média escala, buscando facilitar o uso de aprendizado de máquina para não especialistas, usando uma linguagem de alto nível para propósito geral.(Pedregosa et al., 2011).
   Sendo assim, o scikit será fundamental para os nossos exercícios, visto que ele cria uma camada de abstração para os algoritmos de aprendizagem de máquina, fazendo com que nos preocupemos no conceito dos algoritmos e sua aplicação, e não necessariamente à implementação dos mesmos.


   Classificador Naive Bayes
    
   O classificador Naive Bayes é um método designado para classificação de dados através do aprendizado indutivo supervisionado, que busca performar uma predição correta das classes de testes, possuindo uma abordagem simples e com conhecimento probabilístico. (John & Langley, 1995)
   Além disso, este classificador é um dos mais eficientes em processamento e precisão em relação à rotulação de novas amostras.(Schmitt, 2013).

   Implementação do código

   Visto que já temos um conhecimento básico acerca dos conceitos, algoritmos e ferramentas que vamos utilizar, começaremos agora a implementar um código que, através do algoritmo Gaussian Naive Bayes disponibilizado pelo Scikit Learn, consiga predizer se uma pessoa possui Diabetes ou não à partir de um conjunto de dados de saúde da pessoa.


   Primeiramente, vamos importar as bibliotecas necessárias

```python
import pandas as pd #Necessária para manipular o dataset
import numpy as np
import matplotlib as plt
from sklearn.naive_bayes import GaussianNB #Importando o algoritmo Naive Bayes
```

   Para criarmos o nosso modelo de classificador  Naive Bayeisiano Gaussiano é bem simples, como podemos ver, com apenas uma linha conseguimos criar uma instância do modelo para posteriormente treiná-lo e utilizá-lo para realizar rotulações de novos dados.


```python
#Criando uma instância do modelo Gaussian Naive Bayes e atribuindo ele
#à váriavel model
model = GaussianNB()
```

   Agora iremos abrir nosso conjunto de dados sobre diversas pessoas que possuem diabetes ou não. Para isso, iremos fazer como foi visto na aula sobre Pandas, abrindo o arquivo "diabetes.csv" e criando um dataframe chamado dados com este dataset.


```python
dados = pd.read_csv("diabetes.csv")
dados
```

   Como se pode ver, temos 768 linhas, cada uma representando uma pessoa, e nove colunas, que representam as características que irão ser utilizadas na inferência para predizer se uma pessoa possui diabetes ou não. No nosso caso, a coluna Outcome está representando o dado que queremos predizer, sendo assim, caso seja 0 a pessoa não possui diabetes, caso seja 1 ela possui esta condição.


   Para podermos começar o treino do nosso modelo, precisamos separar os dados em nosso dataframe, separando a coluna que queremos predizer das demais. Para isso vamos realizar uma pequena manipulação dos nomes das colunas do nosso dataframe, como será visto a seguir.


```python
#separa o nome do dado/coluna que vai ser predizido
coluna_alvo=['Outcome'] 
#criamos uma lista com os nomes das colunas/dados que são utilizados para
#a inferencia do dado que iremos predizer, ou seja, dados independentes.
preditores = list(set(list(dados.columns))-set(coluna_alvo)) 
```

   Como podemos ver, através das variáveis preditores e coluna_alvo, que são listas com os nomes das colunas, podemos acessar mais fácil e intuitivamente o dataframe para manipulá-lo mais a seguir.


```python
dados[preditores]
```

```python
dados[coluna_alvo]
```

   A partir de agora, iremos pegar os valores contidos no dataframe para que possamos aplicar no modelo para treiná-lo. Iremos separar em X, todos os valores que serão necessários para realizar a predição da diabetes de alguma pessoas. Já em Y, será justamente os valores ( 0  ou 1) que definem se a pessoa tem diabetes ou não.


```python
X= dados[preditores].values
X
```

```python
Y= dados[coluna_alvo].values
Y
```

   Como pode ser visto, geramos vetores com uma quantidade de elementos igual ao número de linhas do dataframe, sendo cada elemento do vetor, outro vetor, porém com os dados de cada coluna.


   Agora iremos importar um método disponível no SK Learn, que nos permite separar nossos dados em variáveis, para realizarmos o treino e teste do nosso modelo.Dessa forma, teremos um X_train e X_teste, que são subconjuntos de X visto anteriormente, e, de forma análoga, y_train e y_test.
    A função train_test_split possui alguns parâmetros, para ver todos, veja a documentação disponível no site do Scikit Learn. No nosso código, como podemos ver abaixo, estamos usando X (conjunto de variáveis independentes), Y (conjunto dos dados que queremos obter através das predições), test_size para configuração da divisão ( 0.3 implica que serão 70% de treino e 30% de teste)
    Por fim, o random_state funciona da seguinte forma:
        -Caso você não especifique um número, toda vez que você treinar o modelo, ele irá gerar um resultado diferente.
        -Caso você especifique, a "configuração" daquele treino realizado ficará atrelado à aquele número, ou seja, toda vez que você usar um número ele irá reproduzir a mesma configuração.


```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size=0.3, random_state=0)
```

   Agora, com os dados devidamente separados, utilizaremos o método fit() para treinarmos o modelo, como é visto abaixo:

```python
#Utilizamos o método fit em nosso modelo "model", passando como parametro
#Os dados de treino
model.fit(X_train,y_train)
```

   Com o modelo treinado, pode-se agora realizar predições sobre dados novos. Para isto, vamo utilizar o método predict(), que recebe como parâmetro os dados que queremos predizer. Sendo assim, ao utilizarmos esse método, ele irá nos retornar um conjunto de valores, no mesmo formato de y_train ou y_test, então iremos guardar em uma nova variável para realizarmos avaliações métricas.


```python
#Aplicando o teste e guardando o resultado em y_pred
y_pred = model.predict(X_test)
```

   Pronto, com tudo isto feito, já temos nosso modelo treinado e testado, sendo assim, uma boa opção é realizar análises sobre os resultados obtidos. Uma simples métrica de eficiência seria ver a acurácia obtida pelo nosso modelo.Através do metrics podemos verificar isso facilmente, como se pode ver abaixo.


```python
#Importando a biblioteca de análise métrica.
from sklearn import metrics
#Através do método accuracy_score(), podemos passar como parametro o nossos
#dados que foram separados para o teste e os valores que foram obtidos através
#da predição do modelo
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

   Caso se deseje uma visão mais detalhada dos resultados obtidos, existe um outro método, também contida em metrics, que nos gera uma tabela com outros escores de eficiência, como precision, recall e f1-score. Também podemos montar uma matriz de confusão, Vejamos abaixo:


```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```
